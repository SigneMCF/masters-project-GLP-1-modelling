# Normal model with no receptor mediated clearance

```python
reservoir1=Glp_pde()
reservoir1.l_i0=100  #nM
reservoir1.P.L=2000000    #2000.000 #nmeter
reservoir1.P.K_D=1
reservoir1.P.K_D_poor=100 #affinity corresponding to internalization , fraction 0.9 
reservoir1.P.decay=50*reservoir1.P.decay #Reference value is a rate corresponding to T_1/2=10 h 
reservoir1.P.decay_1=0 #no receptor mediated clearance
reservoir1.set_boundary("reservoir")
reservoir1.dz0=10
reservoir1.model="simple"
```

```python

reservoir1.solve(T=reservoir1.P.default_T,init_type='soft')
reservoir1.plot(first_index=28,step_index=6)
```
